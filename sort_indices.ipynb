```python
def first_and_last_ten_debt(df, cols):
    for col in cols:
        df[col] = df[col].apply(lambda x: np.round(x, 7))
        tmp = df[df[col] != 0].copy()
        print(col, end='\n\n')
        tmp[['Имя, заемщик', 'Чин', 'Ранг', 'Титул', col]].sort_values(by=col, ascending=True).head(10).to_excel('results/' + col + '_asc_inflow.xlsx')
        tmp[['Имя, заемщик', 'Чин', 'Ранг', 'Титул', col]].sort_values(by=col, ascending=False).head(10).to_excel('results/' + col + '_desc_inflow.xlsx')
        
def first_and_last_ten_cred(df, cols):
    for col in cols:
        df[col] = df[col].apply(lambda x: np.round(x, 7))
        tmp = df[df[col] != 0].copy()
        print(col, end='\n\n')
        tmp[['Имя, кредитор', 'Чин', 'Ранг', 'Титул', col]].sort_values(by=col, ascending=True).head(10).to_excel('results/' + col + '_asc_inflow.xlsx')
        tmp[['Имя, кредитор', 'Чин', 'Ранг', 'Титул', col]].sort_values(by=col, ascending=False).head(10).to_excel('results/' + col + '_desc_inflow.xlsx')
```

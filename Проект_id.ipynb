```python
fio = df['Заемщик']
grp = ['Банкиры Иван Тамес и компания', 'Бахметевы Николай и Анна Федоровичи', 
       'Безсоновы Катерина и Анна Федоровны', 'Волконские', 'Ершов Алексей Васильевич и Катерина Васильевна', 
       'Львовы', 'Масловы', 'Языковы Александр, Катерина, Любовь, Варвара и Анна Дмитриенвичи', 
       'Гагарины Николай и Сергей Сергеевичи', 'Гагарины дети Сергея Сергеевича',
       'Новикова Прасковья Александровна,Новиков Александр Андреевич', 'Машковы Дочери Петра Ивановича Машкова',
       'Татищевы Екатерина и Марья Павловны ', 'Шефферовы Варвара, Анна и Катерина', 'Мансуровы']
di = {}
personal_id = 1
grp_id = 1
for x in fio:
  if x in di:
    continue
  if x in grp:
    di[x] = 'grp_id_' + str(grp_id)
    grp_id += 1
  else:
    di[x] = 'personal_id_' + str(personal_id)
    personal_id += 1
cred = df['Кредитор']
for x in cred:
  if x in di:
    continue
  if x in grp:
    di[x] = 'grp_id_' + str(grp_id)
    grp_id += 1
  else:
    di[x] = 'personal_id_' + str(personal_id)
    personal_id += 1

fio

# df.columns
```
